## Строим графики

Построим график: распределение количества рейсов по моделям самолетов.

```python
model_flight.sort_values(by='flights_amount',ascending=False).plot(
    'model','flights_amount', kind = 'bar', rot = 65, figsize = (10,5),
    title = 'Количество рейсов по моделям самолетов');
```

Абсолютное большинство рейсов приходится на три модели самолётов:  
* Bombardier CRJ-200  
* Cessna 208 Caravan  
* Sukhoi SuperJet-100	

Видно, что преобладают самолёты небольшой вместимости, что свидетельствует о небольшом пассажиропотоке.  
Так же можно отметить, что Sukhoi SuperJet-100 используется намного чаще, чем Airbus A319-100 такой же вместимости. Авиакомпании отдают предпочтение отечественной технике.

Построим график: количество рейсов в распределении по городам.

```python
city_flight.sort_values(by = 'average_flights', ascending = True).plot(
    'city','average_flights', kind = 'barh',figsize = (12,20), 
    title = 'Количество рейсов в день по городам');
```

Абсолютный лидер по количеству рейсов - Москва.  
Также выделяются Санкт-Петербург и Новосибирск, расположишеея на втором и третьем месте соответственно.  
Количество рейсов в день по городам рапределено неравномерно. Выделятся группа крупнейших российских городов с наибольшим количеством рейсов. Также можно выделить группу небольших городов, в которых регистрируется по 1-2 рейса - в основном это удаленные от наиболее населенной  части страны города.

Изучим топ-10 городов по количеству рейсов

```python
city_flight.sort_values(by = 'average_flights', ascending = True).tail(10).plot(
    'city','average_flights', kind = 'barh',figsize = (12,6), xlim = (0,50), 
    title = 'Топ-10 городов по количеству рейсов в день');
```

В отличие от трех лидеров (Москва, Санкт-Петербург и Новосибирск), в остальных городах количество рейсов практически не отличается и составляет от 9 до 11.  
Москва, Санкт-Петербург и Новосибирск являются тремя крупнейшими городами России. Для Новосибирска наверняка также большую роль играет удаленность от других крупных городов (как и для Красноярска).
